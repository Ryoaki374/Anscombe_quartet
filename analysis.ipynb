```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Anscombe's data set
data = {
    "I": {
        "x": np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]),
        "y": np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]),
    },
    "II": {
        "x": np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]),
        "y": np.array([9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]),
    },
    "III": {
        "x": np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]),
        "y": np.array([7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]),
    },
    "IV": {
        "x": np.array([8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0]),
        "y": np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]),
    }
}
```

```python
# 基本統計量の計算と表示
for group, values in data.items():
    x, y = values["x"], values["y"]
    mean_x, mean_y = np.mean(x), np.mean(y)
    var_x, var_y = np.var(x, ddof=1), np.var(y, ddof=1)
    correlation = np.corrcoef(x, y)[0, 1]
    print(f"Group {group}:")
    print(f"  Mean of x: {mean_x:.2f}, Mean of y: {mean_y:.3f}")
    print(f"  Variance of x: {var_x:.2f}, Variance of y: {var_y:.4f}")
    print(f"  Correlation: {correlation:.4f}\n")

# 各グループの散布図をプロット
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.flatten()

for i, (group, values) in enumerate(data.items()):
    x, y = values["x"], values["y"]
    # 散布図
    axes[i].scatter(x, y, label=f"Group {group}", color="blue")
    # 回帰直線
    m, b = np.polyfit(x, y, 1)
    axes[i].plot(np.arange(0,21,1), m * np.arange(0,21,1) + b, color="red", label=f"y={m:.3f}x+{b:.3f}")
    axes[i].set_title(f"Group {group}")
    axes[i].set_xlabel("x")
    axes[i].set_ylabel("y")
    axes[i].set_xlim(2.5,20)
    axes[i].set_ylim(0,13)
    axes[i].legend()

plt.tight_layout()
plt.show()
```
